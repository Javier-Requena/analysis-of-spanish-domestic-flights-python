```python
import pandas as pd
```

```python
renfe_df = pd.read_csv('opendata_renfe_trips.csv')
```

```python
renfe_df.drop(renfe_df.iloc[:, 8:], axis = 1, inplace = True)
renfe_df.drop(renfe_df.iloc[:, 0:2], axis = 1, inplace = True)
```

```python
renfe_df[['origin', 'destination']] = renfe_df[['origin', 'destination']].applymap(lambda s: str(s).title())
```

```python
renfe_stats = renfe_df.groupby(['origin', 'destination']).describe()
```

```python
renfe_stats.to_csv('renfe_database_statistics.csv')
```
